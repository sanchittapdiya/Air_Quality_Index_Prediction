AQI data is collected from a paid third party app , it have AQI of 2013 to 2018 for every hour
we have take PM 2.5 feature from it

Also, the data we have collected is on daily basis, so we have to convert data from hours to day

we have to combine this data and data from html into one csv file

We have to handle the missing values , invalid data and power fails

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def avg_data_2013():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2013.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average

def avg_data_2014():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2014.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average

def avg_data_2015():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2015.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average

def avg_data_2016():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2016.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average

def avg_data_2017():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2017.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average

def avg_data_2018():
    temp_i=0
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2018.csv',chunksize=24): 
        #chunksize=24 -> represents how many records we are taking at a time, here 24 because of 24 hours a day.
        add_var=0
        avg=0.0
        data=[]
        df=pd.DataFrame(data=rows) #converting list to dataframe
        for index,row in df.iterrows(): #iterating through rows
            data.append(row['PM2.5']) #taking/appending pm2.5 data to data
        #now data have all the values in pm2.5, also it have missing values and invalid datas
        for i in data:
            if type(i) is float or type(i) is int:
                add_var=add_var+i
            elif type(i) is str:
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                    temp=float(i)
                    add_var=add_var+temp
        avg=add_var/24
        temp_i=temp_i+1
        average.append(avg)
    return average
```
